### Question 1: Simple Linear Regression 

Let's set up our workspace and use the `Boston` dataset in the `ISLP` library. Print `Boston` to learn more about the dataset.

```python
# Import standard libraries
import numpy as np
import pandas as pd
import seaborn as sns
from matplotlib.pyplot import subplots
import statsmodels.api as sm
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Import specific objects
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier

from statsmodels.stats.outliers_influence \
     import variance_inflation_factor as VIF
from statsmodels.stats.anova import anova_lm

from ISLP import load_data
from ISLP.models import (ModelSpec as MS,
                         summarize,
                         poly)
```

```python
# Load the "Boston" dataset using the "load_data" function from the ISLP package
Boston = load_data('Boston')
```

Before we fit and review model outputs, we should visualize our data. Review the code and plot, shown below. Answer the following questions:

_(i)_ What are the `medv` and `dis` variables being plotted? (Hint: review this [link](https://islp.readthedocs.io/en/latest/datasets/Boston.html))

We are looking at the relationship between median home values and distance to employment centers in Boston. 

medv = median value of owner-occupied homes in $1000s.
dis = weighted mean of distances to five Boston employment centres.

_(ii)_ What concept ‘defines’ the plotted line?

Linear regression.

```python
# Extract the variables
medv = Boston['medv'].values.reshape(-1, 1)
dis = Boston['dis'].values.reshape(-1, 1)

# Plot data
plt.scatter(dis, medv, label='Data')
plt.xlabel('dis')
plt.ylabel('medv')

# Fit a linear regression model
lm = LinearRegression()
lm.fit(dis, medv)

# Plot the regression line
plt.plot(dis, lm.predict(dis), color='red', label='Regression Line')

# Add labels and legend
plt.xlabel('dis')
plt.ylabel('medv')
plt.legend()

# Show the plot
plt.show()
```

Consider the variables plotted above. In the context of the `Boston` dataset:

### Question 2: Multiple Linear Regression 

We'll continue to use the `Boston` dataset for this question.

_(i)_ Fit a multiple linear regression, with two predictor variables: $X_1$ is `dis`, and $X_2$ is `rm`. As before, keep `medv` as the response variable Y. (Hint: use the syntax `sm.add_constant` with the appropriate arguments).

```python
#Extract the variables
medv = Boston['medv'].values.reshape(-1, 1)
dis = Boston['dis'].values.reshape(-1, 1)
rm = Boston['rm'].values.reshape(-1, 1)

# Plot data
plt.scatter(dis, medv, label='Data')
plt.xlabel('dis')
plt.ylabel('medv')

# Fit a linear regression model for distance to employment centers
lm = LinearRegression()
lm.fit(dis, medv)

# Plot the regression line for distance to employment centers
plt.plot(dis, lm.predict(dis), color='red', label='Regression Line (dis)')

# Fit a linear regression model for average number of rooms
lm2 = LinearRegression()
lm2.fit(rm, medv)

# Plot the regression line for average number of rooms
plt.plot(rm, lm2.predict(rm), color='blue', label='Regression Line (rm)')

# Add labels and legend
plt.xlabel('dis & rm')
plt.ylabel('medv')
plt.legend()

plt.show()
```

```python
X = sm.add_constant(np.column_stack((dis, rm)))

model = sm.OLS(medv, X)
results = model.fit()

print(results.summary())

confidence_intervals = results.conf_int()
print(confidence_intervals)
```

_(ii)_ In the context of the `Boston` dataset, state the null and alternative hypotheses.

The implied null hypothesis is that there is no significant relationship between the distance to employment centers, median home prices and the average number of rooms per dwelling. 

The implied alternative hypothesis is that there is a significant relationship between the distance to employment centers, median home prices and the average number of rooms per dwelling. 

_(iii)_ Review the model output, using `summary()`. Does it appear that both `dis` and `rm` are predictive of `medv`? How did you determine this?

In reviewing the coefficient, p-values and $R^2$ values I believe both dis and rm are predictive of medv. 

As the p-values are less than 0.05 this indicates that dis and rm are statistically significant predictors of medv. Additionally, each time dis increases rm increases by 8.8014. In this case the $R^2$ value indicates 49.6% of the variance in the dependent variable is explained by the independant variables.

_(iv)_ We can use the inbuilt `sm.graphics.plot_regress_exog` function to generate helpful diagnostic plots (Hint: provide `plot_regress_exog` with the multiple regression model). Review the first generated plot, 'Residuals vs. Fitted'. Which observations are outliers? What impact might outliers have on our model?

In reviewing the residuals vs. fitted graph, points that are further from the main cluster can be considered as outliers. 

Outliers influence the model fit, prediction accuracy and coefficients. 

```python
medv = Boston['medv']
dis = Boston['dis']
rm = Boston['rm']

X = pd.DataFrame({'dis': dis, 'rm': rm})

X = sm.add_constant(X)

model = sm.OLS(medv, X).fit()

sm.graphics.plot_regress_exog(model, 'dis')
plt.show()
```

_(v)_ Fit a second model, this time including an interaction between the two predictor variables. Is there an interaction? (Hint: add a variable `x1 * x2` where `x1` and `x2` are the predictor variables). State an interpretation of the interaction, in the context of the `Boston` dataset, in one or two sentences.

In the context of the Boston dataset, I believe the distance to employment centers and median home values is depandant on the number of rooms in a house.

The coefficents indicate that as distance to employment centers increases the median house value goes down. Each addition of a room leads to the median home value to go up. The interaction coeffiecnt indicates that when both distance to employment centers and the number of rooms increase together, the effect of distance on median home value increases.


```python
interaction = dis * rm

X_interact = pd.DataFrame({'dis': dis, 'rm': rm, 'interaction': interaction})

X_interact = sm.add_constant(X_interact)

model_interact = sm.OLS(medv, X_interact).fit()

print(model_interact.summary())
```
